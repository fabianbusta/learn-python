## Problema

¿Como identificar a un número primo?

- Un número natural p > 1 es primo si y sólo si, sus únicos factores son exactamente p y 1.
- En otras palabras los numeros primos son aquellos enteros positivos mayores que 1 y solo son divisibles por sí mismos y por 1.

## Solución

Método fuerza bruta.

To-do

1. Solicitar al usuario que ingrese un número entero positivo.
2. Verificar si el numero es mayor que 1.
3. Si el número es mayor que 1:
   1. Inicializar un divisor en 2.
   2. Inicializar una variable booleana llamada es_primo en Verdadero.
   3. Mientras el divisor sea menor que el número:
      1. Verificar si el número es divisible por el divisor.
      2. Si es divisible, cambiar es_primo a Falso y salir del bucle.
      3. Incrementar el divisor en 1.
   4. Si es_primo es Verdadero, imprimir mensaje Es primo.
   5. De lo contrario, imprimir mensaje No es primo, 
4. Si el número es menor o igual que 1, imprimir un mensaje de error

```python
# Solicitar al usuario ingresar un número entero positivo
numero = int(input("Ingresa un número entero positivo: "))

# Verificar si el numero es mayor que 1
if numero > 1:
    # Si el número es mayor que 1
    # Inicializar divisor en 2
    divisor = 2
    # Inicializar una variable booleana en Verdadero
    es_primo = True
    
    # Mientras el divisor sea menor que el número
    while divisor < numero:
        # print(f"Chequeando: {numero}/{divisor} = {numero/divisor}") # Ver en cada iteración, solo para pruebas
        
        # Verificar si el número es divisible por el divisor
        if numero % divisor == 0:
            # Si es divisible, cambiar es_primo a Falso y salir del bucle
            es_primo = False
            break
        # Incrementar el divisor en 1.
        divisor += 1
    
    if es_primo:
        # Si es_primo es Verdadero, imprimir mensaje
        print(f"{numero} es un número primo.")
    else:
        # De lo contrario 
        print(f"{numero} no es un número primo.")
else:
    print("Por favor, introduce un número entero positivo")
```
